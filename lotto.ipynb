```python
import random
import time

# 1. 중복을 체크하며 6개를 뽑는 함수 (추첨만 담당)
def get_lotto_raw_data():
    lotto_list = []
    print("번호 추첨기를 가동합니다")
    print("-" * 35)

    while len(lotto_list) < 6:
        new_num = random.randint(1, 45)

        # 중복되지 않은 경우에만 화면에 보여주고 리스트에 추가
        if new_num not in lotto_list:
            lotto_list.append(new_num)
            print(f"새로운 번호 추출: {new_num}")
            time.sleep(1) # 코랩에서 추출 과정을 보기 위한 딜레이
        else:
            print(f"중복 발생 ({new_num}): 다시 뽑습니다.")

    # 정렬되지 않은 상태 그대로 리스트 반환
    return lotto_list

# 2. 함수 호출 및 결과값(return) 받기
raw_numbers = get_lotto_raw_data()

# 3. 함수 밖에서 정렬 및 결과 보고 수행
raw_numbers.sort()  # 오름차순 정렬

print("-" * 35)
print(f"[최종 결과 보고]")
print(f"금주의 행운 번호는 {raw_numbers} 입니다!")
print("-" * 35)
```
